```python
# Get the RGB of a given pixel
# TOCOMPLETE - DONE
pixel = my_image.getpixel((120, 60))
print("RBG of pixel:", pixel)
print("R :", pixel[0])
print("G :", pixel[1])
print("B :", pixel[2])
```

```python
# Pull all pixels in black
for x in range(w):
  for y in range(h):
     pixel = my_image.getpixel((x,y))
     my_image.putpixel((x,y), (0,0,0))
my_image
```

```python
# Put all pixels for x and y are even in black
# RGB = 0,0,0
for x in range(w):
  if x % 2 == 0:
    for y in range(h):
      if y % 2 == 0:
          pixel = my_image.getpixel((x,y))
          my_image.putpixel((x,y), (0,0,0))

  # or, put the if-statements together after the y for-loop
my_image
```

```python
# Pixel average
# TOCOMPLETE
def pixel_average(p):
  return int((p[0] + p[1] + p[2]) / 3) # (R+G+B)/3
```

```python
# Replace bright pixels by grey pixels
# bright if the pixel average > 153 (60% max brightness)
# TOCOMPLETE
def filter_average(img):
  for x in range(img.width):
    for y in range(img.height):
      pixel = img.getpixel((x,y))
      pixels_average = pixel_average(pixel)
      if pixels_average > 153:
        img.putpixel((x,y),(pixels_average,pixels_average,pixels_average))

filter_average(my_image)
my_image
```

```python
def my_filter(img):
  for x in range(img.width):
    for y in range(img.height):
      pixel = img.getpixel((x,y))
      pixels_average = pixel_average(pixel)
      if pixels_average >= 130:
        img.putpixel((x,y), (x+100,y+100,203))

my_filter(my_image)
my_image
```
